```python
import matplotlib.pyplot as plt
import pandas as pd
mtcars=pd.read_csv('/content/mtcars11.csv')
trans_count=mtcars['am'].value_counts()
trans_count.plot(kind='bar')
plt.title("transmission of frequency")
plt.xlabel("transmission'(0=auto,1=manual)")
plt.ylabel("count")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
mtcars=pd.read_csv('/content/mtcars11.csv')
sns.scatterplot(x='wt',y='mpg',data=mtcars)
plt.title("weight vs mpg")
plt.xlabel("weight")
plt.ylabel("mpg")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
mtcars=pd.read_csv('/content/mtcars11.csv')
import seaborn as sns
sns.boxplot(y=mtcars['mpg'])
plt.title("boxplot of mpg")
plt.ylabel("mpg")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
mtcars=pd.read_csv('/content/mtcars11.csv')
import seaborn as sns
plt.hist(mtcars["mpg"],bins=10,color='pink',edgecolor='black')
plt.xlabel("mpg")
plt.ylabel("frequency")
plt.title("distribution")
plt.show()
```
